```python
%matplotlib inline
from scipy.fftpack import fft, ifft, fftshift, ifftshift
from scipy.io import wavfile
from scipy.signal import ricker
from sympy.mpmath import sech
import math
import matplotlib.pyplot as plt
import numpy as np
import random
import string
```

##Time Frequency Analysis - Windowed Fourier Transforms

Fourier transform based methods have a limitation in that while it captures all the frequency content of a signal, it fails to capture the moment in time when these frequencies were exhibited (since by definition, time is integrated out). In the context of signal processing, in order to apply FFTs directly, we have to assume that the signal has a fixed (stationary or periodic) frequency.

In order to analyze signals whose frequency values change over time (ie non-stationary signals), we need to extract both time and frequency signals. The simplest approach to doing this is to decompose the signal over the time domain into separate time frames, and FFT applied to each window to characterize the frequencies present during that time frame. The code below splits up a section of a sound wave into 3 windows, then applies the FFT to each - this gives us the ability to extract both time and frequency information from the signal.

```python
# Grab some sound from file (from Caleb Madrigal's talk on Fourier Analysis at OSCON)
# and analyse a small slice
fs, sig = wavfile.read("../data/doremi.wav")
ys = sig[120000:120600]
xs = range(len(ys))

ws = [(x, x+200) for x in range(0,600,200)]

plt.subplot(4, 1, 1, title="(a)")
plt.xlabel("time")
plt.ylabel("S(t)")
plt.plot(xs, ys)
for w in ws:
    plt.axvline(w[0], color='r', linestyle='--')

for i in range(len(ws)):
    ut = fft(ys[ws[i][0]:ws[i][1]])
    plt.subplot(4, 1, i+2, title="(%s)" % string.lowercase[i+1])
    plt.xlabel("frequency")
    plt.ylabel("|FFT(S)|")
    plt.plot(range(len(ut)), np.abs(fftshift(ut)) / np.max(np.abs(fftshift(ut))))

fig = plt.gcf()
fig.set_size_inches(6, 6)
plt.tight_layout()
```

Gabor proposed a formal method for localising both time and frequency - the Gabor Tranform, also known as the Short Time Fourier Transform (STFT). It acts as a time filter for localizing the signal over a specific window of time. The window is slid down the entire signal and picks up the frequency information at each instant of time.

There is a resolution trade-off with the Gabor Transform. As the filtering window is increased, increased frequency resolution is gained at the expense of decreased time resolution, and vice versa.

The code below attempts to reconstruct a signal using Gabor transforms with window size of 5, 25 and 50 respectively and a hop size of 5. The incoming signal is first chopped up into overlapping windows of the specified window size and its FFT taken. The FFTs are averaged (with offset) and the resulting average in the frequency domain converted back into the time domain. As the time window is increased, the curve gets more complex (indicating more granular frequencies), but looks less and less like the original. 

```python
def apply_gabor_transform(a, hop, xs, ys):
    window = (0, a)
    target = np.zeros((len(xs)))
    while True:
        ut = fft(xs[window[0]:window[1]])
        target[window[0]:window[1]] += ut
        new_start = window[0] + hop
        if new_start + a > len(xs):
            break
        window = (new_start, new_start + a)
    return ifft(target)
        
ts = np.linspace(-6*math.pi, 6*math.pi, 100)
u = 5*np.sin(ts) + 4*np.sin(2*ts) + 3*np.sin(3*ts) + 2*np.sin(4*ts)

# plot signal in time domain
plt.subplot(4, 1, 1, title="(a)")
plt.plot(ts, u)

avals = [5, 25, 50]
sid = 2
for i in range(len(avals)):
    plt.subplot(4, 1, sid, title="(%s)" % string.lowercase[sid-1])
    ug = apply_gabor_transform(avals[i], 5, ts, u)
    plt.plot(ts, np.abs(ifftshift(ug)))
    sid += 1
    
fig = plt.gcf()
fig.set_size_inches(8, 8)
plt.tight_layout()
```

##Wavelet Transforms - the Haar Wavelet

A modification of the Gabor Transform is to allow the scaling window to vary according to the frequency being extracted. First high frequency components are extracted using a narrow scaling window. The window is successively increased to extract lower frequencies (and poorer time resolution). This is the fundamental principle of wavelet theory.

The first Wavelet we will consider is the Haar wavelet. The wavelet can be represented by the following equation:

&psi;<sub>a,b</sub>(t) = (1/&radic;a) &psi;((t - b) / a)

&psi;(t) = 1, -1 or 0 if 0 <= t <= 0.5; 0.5 <= t < 1; otherwise

Its a square wave as shown below.

```python
def haar_wavelet(a, b, t):
    x = 1.0 * (t - b) / a
    while x > 1:
        x = x - 1
    if x >= 0 and x < 0.5:
        return -1.0 / math.sqrt(a)
    elif x >= 0.5 and x < 1:
        return 1.0 / math.sqrt(a)
    else:
        return 0.0

ts = np.linspace(0, 12*math.pi, 128)
i = 1
for a in [5, 10, 20]:
    ys = [haar_wavelet(a, 0, t) for t in ts]
    plt.subplot(3, 1, i, title="a = %d" % (a))
    plt.plot(ts, ys)
    i += 1
plt.tight_layout()
```

The code below demonstrates how the Haar wavelet would be used to reconstruct a signal back into its components. Like the Gabor transform, we use a sliding window, except the window now contains the Haar wavelet. The sliding window with the smallest a value (width) is applied to the signal in the time domain, and the Haar wavelet filter is applied to the FFT of the slice. This slice is added to the reconstructed signal and removed from the input signal slice. The operation is repeated for each slice. Then the same operation is repeated on all slices, filtering using a Haar wavelet with the next width (double the previous one), until the width spans the full width of the input signal.

The output shows (a) the original signal, and (b-d) the reconstructed signal with Haar wavelet filters of widths 5, 25 and 50 respectively.

```python
def apply_haar_wavelet(ts, u, a, hop):
    utarget = np.zeros((ts.shape[0]))
    for i in range(int(math.log(len(ts), 2)) + 1):
        window = (0, a)
        while True:
            ut = fft(u[window[0]:window[1]])
            w = np.array([haar_wavelet(2**i, 5, t) for t in ts[window[0]:window[1]]])
            uth = fft(np.multiply(w, ut))
            utarget[window[0]:window[1]] += uth
            ut = ut - uth
            new_start = window[0] + hop
            if new_start + a > len(u):
                break
            window = (new_start, new_start + a)
    return ifft(utarget)
    
ts = np.linspace(-6*math.pi, 6*math.pi, 128)
u = 5*np.sin(ts) + 4*np.sin(2*ts) + 3*np.sin(3*ts) + 2*np.sin(4*ts)

# plot signal in time domain
plt.subplot(4, 1, 1, title="(a)")
plt.plot(ts, u)

avals = [5, 25, 50]
for i in range(len(avals)):
    u_recon = apply_haar_wavelet(ts, u, avals[i], 5)
    plt.subplot(4, 1, i+2, title="(%s)" % string.lowercase[i+1])
    plt.plot(ts, ifftshift(u_recon))
fig = plt.gcf()
fig.set_size_inches(8, 8)
plt.tight_layout()
```

##Wavelet Transforms - the Mexican Hat Wavelet

The Mexican Hat Wavelet is a slightly more complex wavelet with (experimentally observed) better signal reconstruction properties, especially with seismic data. The formula for the Mexican Hat Wavelet is as follows:

&psi;(t) = (1 - 2&pi;<sup>2</sup>a<sup>2</sup>t<sup>2</sup>)e<sup>-2&pi;<sup>2</sup>a<sup>2t<sup>2</sup>/2</sup>

```python
ts = np.linspace(0, 12*math.pi, 128)

i = 1
for a in [5, 10, 20]:
    ys = ricker(len(ts), a)
    plt.subplot(3, 1, i, title="a = %d" % (a))
    plt.plot(ys)
    i += 1
plt.tight_layout()
```

We apply this wavelet as before to our signal. The logic is identical to that used for reconstructing the signal using the Haar wavelet, except we use the canned ricker() method in scipy.signals for generating the Mexican Hat wavelet. As before, the (a) represents the original signal, and (b-d) represents the reconstructed signal using a sliding wavelet of width 5, 25 and 50. 

```python
def apply_mexican_hat_wavelet(ts, u, a, hop):
    utarget = np.zeros((ts.shape[0]))
    for i in range(int(math.log(len(ts), 2)) + 1):
        window = (0, a)
        while True:
            ut = fft(u[window[0]:window[1]])
            w = ricker(len(ts[window[0]:window[1]]), a)
            uth = fft(np.multiply(w, ut))
            utarget[window[0]:window[1]] += uth
            ut = ut - uth
            new_start = window[0] + hop
            if new_start + a > len(u):
                break
            window = (new_start, new_start + a)
    return ifft(utarget)
    
ts = np.linspace(-6*math.pi, 6*math.pi, 128)
u = 5*np.sin(ts) + 4*np.sin(2*ts) + 3*np.sin(3*ts) + 2*np.sin(4*ts)

# plot signal in time domain
plt.subplot(4, 1, 1, title="(a)")
plt.plot(ts, u)

avals = [5, 25, 50]
for i in range(len(avals)):
    u_recon = apply_mexican_hat_wavelet(ts, u, avals[i], 5)
    plt.subplot(4, 1, i+2, title="(%s)" % string.lowercase[i+1])
    plt.plot(ts, ifftshift(u_recon))
fig = plt.gcf()
fig.set_size_inches(8, 8)
plt.tight_layout()
```
