```python
%matplotlib inline
from scipy.fftpack import fft, ifft, fftshift, ifftshift
from sympy.mpmath import sech
import math
import matplotlib.pyplot as plt
import numpy as np
import random
import string
```

##Time-Frequency Analysis - Fourier and Wavelet Transforms

Fourier transforms, and more generally spectral transforms, is a very efficient technique for solving a wide variety of problems in physics, biology and engineering. The key idea of the Fourier transform is to represent functions and their derivatives as sums of cosines and sines.

The Fast Fourier Transform (FFT) is an O(N log N) process, more efficient than other linear solvers that are O(N<sup>2</sup>). Given a function that has been discretized into 2<sup>n</sup> points and represented by the vector <b>x</b>, the FFT is found with the function fft(<b>x</b>).

The code below demonstrates the various functions in the FFT family on a Gaussian function f(x) = e<sup>-&alpha;x<sup>2</sup></sup>. The first plot represents the original function with &alpha;=1, the second one shows the FFT of this function, the third one shows the centered version of the FFT and the fourth one represents FFT transformed back to the original space.

```python
l = 20   # define computational domain (-l/2, l/2)
n = 128  # define number of Fourier modes 2**n
xs = np.linspace(-l/2, l/2, n)
ks = [(2 * math.pi * x / l) for x in np.linspace(-n/2, n/2, n)]

u = [math.exp(-x * x) for x in xs]
ut = fft(u)
utshift = fftshift(ut)
ur = ifft(ifftshift(utshift))

plt.subplot(2, 2, 1, title="(a)"); plt.plot(xs, u)
plt.subplot(2, 2, 2, title="(b)"); plt.plot(ks, ut)
plt.subplot(2, 2, 3, title="(c)"); plt.plot(ks, utshift)
plt.subplot(2, 2, 4, title="(d)"); plt.plot(xs, ur)
fig = plt.gcf()
fig.set_size_inches(6, 6)
plt.tight_layout()
```

##Radar Detection and Filtering

Radars work by emitting an electromagnetic field of fixed frequency and listening for reflections of this field from objects in the environment and trying to determine what these objects are. This is complicated by other sources of radiation, extraneous objects, etc, which combine to create a noisy signal for the detector.

One approach to denoising is called filtering. Since noise usually affects all frequencies in the same way, we can simulate noise by transforming the original signal into the frequency domain and adding random noise to it. The code below converts the original signal to the frequency domain using the FFT, adds different levels of noise and converts it back to the time domain. For low noise levels, the signal is still detectable, but gets progressively more difficult as noise levels increase.

```python
l = 60   # time slot to transform
n = 512  # number of Fourier modes (2**9)
ts = np.linspace(-l/2, l/2, n)
ks = [(2 * math.pi * x / l) for x in np.linspace(-n/2, n/2, n)]

u = [sech(t) for t in ts]
plt.subplot(3, 1, 1, title="(%s)" % (string.lowercase[0])); plt.plot(ts, u)

# add noise
noise = [5, 30]
for i in range(len(noise)):
    ut = fft(u)
    utn = ut + [noise[i] * (random.random() + 1j*random.random()) for k in range(n)]
    un = ifft(utn)
    plt.subplot(3, 1, i+2, title="(%s)" % (string.lowercase[i+1]))
    plt.axis([-30, 30, 0, 1])
    plt.plot(ts, abs(un))
fig = plt.gcf()
fig.set_size_inches(6, 6)
plt.tight_layout()
```

Filtering can help detect a signal buried in a noisy field. In case of radar detection, we know that we are looking for a certain frequency band, hence spectral filtering around this can remove undesired frequencies and much of the white noise. Many types of filters exist, simplest is the Gaussian filter, given by:

f(k) = e<sup>-&tau;(k - k<sub>0</sub>)<sup>2</sup></sup> 

In this case we want to apply the filter around k<sub>0</sub>=0. The first chart is the normalized original signal in the frequency domain with white noise added and the filter superimposed (in red), the next chart is the filtered signal in the frequency domain, and the final chart is the reconstruction of the signal in the time domain with the noise filtered out, the original signal (in red) and the black dotted line represents a threshold above which the detector would read the extracted signal as the target (TODO: this doesn't seem to work as advertized).

```python
def apply_filter(u, utn, k0):
    gfilt = np.array([math.exp(-0.2*(k - k0)**2) for k in ks])
    utnf = np.multiply(gfilt, utn)
    unf = ifft(utnf)

    # plot of noisy frequency and gaussian filter
    plt.subplot(3, 1, 1, title="(a)")
    plt.ylim([0, 1])
    plt.xlabel("wavenumber (k)")
    plt.plot(ks, np.abs(utn) / np.max(np.abs(utn)))
    plt.plot(ks, gfilt, 'r-')
    
    # plot of filtered frequency
    plt.subplot(3, 1, 2, title="(b)")
    plt.ylim([0, 1])
    plt.xlabel("wavenumber (k)")
    plt.plot(ks, np.abs(utnf) / np.max(np.abs(utnf)))

    # plot of reconstructed signal in time domain, with original signal (red) and
    # decision threshold (black dashed line)
    plt.subplot(3, 1, 3, title="(c)")
    plt.ylim([0, 1])
    plt.xlabel("time (t)")
    plt.plot(ts, np.abs(unf))
    plt.plot(ts, np.abs(u), 'r-')
    plt.plot(ts, np.ones((n))*0.5, 'k--')
    
    fig = plt.gcf()
    fig.set_size_inches(8, 8)
    plt.tight_layout()
    

apply_filter(u, utn, 0)
```

In contrast, moving k0 to 15 where (we know) there is no signal shows a reconstructed signal that is much below the decision threshold at this location.

```python
apply_filter(u, utn, 15)
```

##Radar Detection and Averaging

The averaging technique uses the idea that white noise can be modeled as a normal random variable with mean 0 and standard deviation 1 added to each frequency of the incoming signal. Thus over many signals, the white noise should add up to 0.

The code below demonstrates this by adding up the same signal with random noise added in the frequency domain and averaging the results. The resulting average signal is transformed back to the time domain and compared with the original signal. The code demonstrates this process for averaging 1, 2, 5 and 100 noisy signals. As can be seen the reconstructed signal gets closer and closer to the original signal (superimposed in red) as the number of signals that are averaged on increases.

```python
def average_signal(u, nrs, noise):
    ut = fft(u)
    averaged = np.zeros((nrs, len(u)))
    for i in range(nrs):
        utn = ut + [noise * (random.random() + 1j*random.random()) for k in range(len(u))]
        averaged[i, :] = np.array(utn)
    return np.mean(averaged, axis=0)
    
l = 60   # time slot to transform
n = 512  # number of Fourier modes (2**9)
ts = np.linspace(-l/2, l/2, n)
ks = [(2 * math.pi * x / l) for x in np.linspace(-n/2, n/2, n)]
noise = 20

u = [sech(t) for t in ts]
ut = fft(u)

num_realizations = [1, 2, 5, 100]
for i in range(len(num_realizations)):
    uta = average_signal(u, num_realizations[i], noise)
    
    plt.subplot(len(num_realizations), 1, i+1, title="(%s)" % (string.lowercase[i]))
    plt.ylim([0, 1])
    plt.plot(ts, np.abs(fftshift(uta)) / np.max(np.abs(uta)))
    plt.xlabel("frequency")
    plt.plot(ts, np.abs(fftshift(ut)) / np.max(np.abs(ut)), 'r')
fig = plt.gcf()
fig.set_size_inches(8, 8)
plt.tight_layout()
```

## Detecting a Time-varying signal

One consequence of the averaging is that we lose the time-domain dynamics. Transferring the average frequency spectrum back to the time domain will only give us the averaged time domain signal, not how it evolved over time. For example, in the case of the radar problem, the movement of the signal represents a moving aircraft. However, regardless of movement of the signal in the time domain, the signal remains fixed in the frequency domain.

The following code demonstrates averaging over a time varying signal in both the time and frequency domains. As can be seen, averaging over the time domain (a) does not yield any information, but averaging over the frequency domain (b) accurately pinpoints the frequency even for a time varying signal.

```python
l = 60   # time slot to transform
n = 512  # number of Fourier modes (2**9)
noise = 30

slices = np.linspace(0, 10, 20)
ts = np.linspace(-l/2, l/2, n)
ks = [(2 * math.pi * x / l) for x in np.linspace(-n/2, n/2, n)]

tsx, ssx = np.meshgrid(ts, slices)
ksx, ssx = np.meshgrid(ks, slices)

def f_sech(x): return sech(x)
np_sech = np.vectorize(f_sech)
U = np_sech(10.0 * np.sin(ssx))

ua = np.zeros(U.shape[1])
uta = np.zeros(U.shape[1])
for i in range(U.shape[0]):
    u = [x for x in U[i, :].tolist()]
    ut = fftshift(fft(u))
    utn = ut + noise * (np.random.randn(1, U.shape[1]))
    un = ifft(ifftshift(utn))
    uta = uta + utn
    ua = ua + un
ua = np.divide(ua, 1.0*U.shape[0])
uta = np.divide(uta, 1.0*U.shape[0])

plt.subplot(2, 1, 1, title="(a)")
plt.xlabel("time")
plt.ylabel("|u|")
plt.plot(ts, np.squeeze(np.abs(ua)))

plt.subplot(2, 1, 2, title="(b)")
plt.xlabel("frequency")
plt.ylabel("|fft(u)|")
plt.plot(ks, np.squeeze(np.abs(uta)))

plt.tight_layout()
```
