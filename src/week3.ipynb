```python
%matplotlib inline
from IPython.display import display, clear_output
from scipy.fftpack import fft, ifft, fftshift, ifftshift
import math
import matplotlib.pyplot as plt
import numpy as np
import random
import string
```

## Spectrograms and Gabor Transforms

A spectrogram represents a signal in both time and frequency domain, sort of a fingerprint of the sound. This will also demonstrate the time vs frequency localization tradeoffs described earlier.

The code below constructs a fairly complicated signal and its FFT and graphs them.

```python
l = 10
n = 2048
ts = np.linspace(0, l, n)
ks = [(2 * math.pi * x / l) for x in np.linspace(-n/2, n/2, n)]

s = 3*np.sin(2*ts) + 0.5*np.tanh(0.5*(ts-3)) + 0.2*np.exp(-(ts-4)**2) + \
    1.5*np.sin(5*ts) + 4*np.cos(3*(ts-6)**2)/10 + (ts/20)**3
st = fft(s)

plt.subplot(2, 1, 1, title="(a)")
plt.xlabel("time")
plt.ylabel("s")
plt.plot(ts, s / np.max(s))

plt.subplot(2, 1, 2, title="(b)")
plt.xlabel("frequency")
plt.ylabel("|fft(s)|")
plt.xlim([-50, 50])
plt.plot(ks, np.abs(fftshift(st)) / np.max(np.abs(fftshift(st))))

plt.tight_layout()
```

The signal s(t) will now be analyzed using the Gabor tranform method. The  simplest Gabor window is a Gaussian time filter centered at some time &tau; with width a. The parameter a determines the time resolution vs the frequency resolution in a time-frequency plot. The narrower the time filtering, the better resolution in time and the worse resolution in frequency.

```python
avals = [10, 1, 0.2]
for j in range(len(avals)):
    g = np.exp(-avals[j]*(ts-4)**2)
    plt.subplot(3, 1, j+1)
    plt.plot(ts, s / np.max(s))
    plt.plot(ts, g, 'r-')
plt.tight_layout()
```

We now multiply the Gabor filter with the original signal S(t) in order to produce a windowed section of the signal. We use a=2 and b=4 for our Gabor filter.

```python
g = np.exp(-2*(ts-4)**2)
sg = np.multiply(g, s)
sgt = fft(sg)

plt.subplot(3, 1, 1)
plt.xlabel("time")
plt.ylabel("S(t), g(t)")
plt.plot(ts, s / np.max(s))
plt.plot(ts, g, 'r-')

plt.subplot(3, 1, 2)
plt.xlabel("time")
plt.ylabel("Sg")
plt.plot(ts, sg)

plt.subplot(3, 1, 3)
plt.xlabel("frequency")
plt.ylabel("|FFT(Sg)|")
plt.xlim([-50, 50])
plt.plot(ks, np.abs(fftshift(sgt)) / np.max(np.abs(fftshift(sgt))))

plt.tight_layout()
```

The key to generating a spectograph is to vary the position b of the time filter and produce spectra at each position in time. The following code generates a movie of this process as the parameter b is translated (TODO: need to be able to clear previous iteration, and remove second copy of chart).

```python
tslide = np.linspace(0, 10, 100)
S = np.zeros((s.shape[0], tslide.shape[0]))

for j in range(tslide.shape[0]):
    g = np.exp(-2*(ts-tslide[j])**2)
    sg = np.multiply(g, s)
    sgt = fft(sg)
    S[:, j] = np.abs(fftshift(sgt))
    
    plt.subplot(3, 1, 1)
    plt.xlabel("time")
    plt.ylabel("S(t), g(t)")
    plt.plot(ts, s / np.max(s))
    plt.plot(ts, g, 'r-')

    plt.subplot(3, 1, 2)
    plt.xlabel("time")
    plt.ylabel("Sg")
    plt.plot(ts, sg)

    plt.subplot(3, 1, 3)
    plt.xlabel("frequency")
    plt.ylabel("|FFT(Sg)|")
    plt.xlim([-50, 50])
    plt.plot(ks, np.abs(fftshift(sgt)) / np.max(np.abs(fftshift(sgt))))
    
    plt.tight_layout()
    plt.pause(0.1)
    clear_output(wait=True)
    display(plt.gcf())
```

The code above also populates a matrix S which contains the Fourier transform of each slice in time of the parameter b. This matrix is used to produce the spectrogram of the time-frequency signal. The following code makes a heatmap visualization of the spectrogram.

```python
plt.ylim([-50, 50])
plt.pcolor(tslide, np.array(ks), S, cmap='RdBu')
```
